```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from xgboost import XGBClassifier
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, roc_curve, confusion_matrix, ConfusionMatrixDisplay
)

# 1. Cargar datos
train = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vS60HQztm8Bh9VS0LiK9Uhr9llcpJnxpefPrNKMC7TjWY3PslbhPBLDuc2Pf2t-yzwO6nXOPGpHSdr-/pub?output=csv")
test = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vTuZODZ6QTsYqzReqNm9arsBIc-tusb063SPYCB6riVdpbezq0lZs3mWDIvYVOodjvM4zPe-QekUS5Q/pub?output=csv")

# 2. Preparar variables
y = train['Subscription']
X = train.drop(columns=['Subscription'])

# Guardar ID del test si tiene (por ejemplo la primera columna)
id_test = test.iloc[:, 0]
X_test = test.copy()

# Identificar variables numricas y categricas
num_features = X.select_dtypes(include=[np.number]).columns.tolist()
cat_features = X.select_dtypes(include='object').columns.tolist()

# 3. Preprocesamiento
numeric_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer([
    ('num', numeric_transformer, num_features),
    ('cat', categorical_transformer, cat_features)
])

# 4. Pipeline del modelo
clf_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42))
])

# 5. Divisin en entrenamiento y validacin
X_train, X_valid, y_train, y_valid = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42)

# 6. Bsqueda de hiperparmetros
param_grid = {
    'classifier__n_estimators': [100, 200, 300],
    'classifier__max_depth': [3, 5, 7],
    'classifier__learning_rate': [0.01, 0.1, 0.2],
    'classifier__subsample': [0.7, 0.9, 1.0],
    'classifier__colsample_bytree': [0.7, 0.9, 1.0]
}

random_search = RandomizedSearchCV(
    clf_pipeline, param_distributions=param_grid,
    n_iter=20, scoring='roc_auc', cv=5, verbose=2, random_state=42
)

random_search.fit(X_train, y_train)
best_model = random_search.best_estimator_

# 7. Interpretacin de la importancia de variables
importances = best_model.named_steps['classifier'].feature_importances_
feature_names = best_model.named_steps['preprocessor'].get_feature_names_out()
feat_importance = pd.Series(importances, index=feature_names).sort_values(ascending=False)

plt.figure(figsize=(8,6))
feat_importance.head(10).plot(kind='barh', color='skyblue')
plt.title("Top 10 variables ms importantes")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

# 8. Evaluacin del desempeo en validacin
y_pred_proba = best_model.predict_proba(X_valid)[:, 1]
y_pred = best_model.predict(X_valid)

auc = roc_auc_score(y_valid, y_pred_proba)
accuracy = accuracy_score(y_valid, y_pred)
precision = precision_score(y_valid, y_pred)
recall = recall_score(y_valid, y_pred)
f1 = f1_score(y_valid, y_pred)
cm = confusion_matrix(y_valid, y_pred)

print(f"AUC-ROC: {auc:.4f}")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=best_model.classes_)
disp.plot(cmap='Blues')
plt.title('Matriz de Confusin')
plt.show()

fpr, tpr, _ = roc_curve(y_valid, y_pred_proba)
plt.plot(fpr, tpr, label=f'AUC = {auc:.4f}')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('Tasa Falsos Positivos')
plt.ylabel('Tasa Verdaderos Positivos')
plt.title('Curva ROC')
plt.legend()
plt.grid(True)
plt.show()

# 9. Prediccin para conjunto de prueba
# Asegurar columnas test coincidan con train
for col in X.columns:
    if col not in X_test.columns:
        X_test[col] = np.nan
X_test = X_test[X.columns]

preds_test = best_model.predict_proba(X_test)[:,1]

submission = pd.DataFrame({
    'Id': id_test,
    'Predicted': preds_test
})
#submission.to_csv('submission.csv', index=False)
#print("Archivo 'submission.csv' guardado correctamente.")


```

```python
def visualizar_metricas_completas(y_val, y_pred, y_pred_proba):
    """
    Visualiza todas las mtricas importantes en subplots
    """
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    
    # 1. Matriz de Confusin
    cm = confusion_matrix(y_val, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, 
                                display_labels=['No Subscription', 'Subscription'])
    disp.plot(ax=axes[0,0], cmap='Blues')
    axes[0,0].set_title('Matriz de Confusin')
    
    # 2. Curva ROC
    fpr, tpr, _ = roc_curve(y_val, y_pred_proba)
    auc_roc = roc_auc_score(y_val, y_pred_proba)
    axes[0,1].plot(fpr, tpr, label=f'AUC = {auc_roc:.4f}', linewidth=2)
    axes[0,1].plot([0, 1], [0, 1], 'k--', alpha=0.7)
    axes[0,1].set_xlabel('Tasa Falsos Positivos')
    axes[0,1].set_ylabel('Tasa Verdaderos Positivos')
    axes[0,1].set_title('Curva ROC')
    axes[0,1].legend()
    axes[0,1].grid(True)
    
    # 3. Curva Precision-Recall
    precision_curve, recall_curve, _ = precision_recall_curve(y_val, y_pred_proba)
    auc_pr = auc(recall_curve, precision_curve)
    axes[1,0].plot(recall_curve, precision_curve, label=f'AUC = {auc_pr:.4f}', linewidth=2)
    axes[1,0].set_xlabel('Recall')
    axes[1,0].set_ylabel('Precision')
    axes[1,0].set_title('Curva Precision-Recall')
    axes[1,0].legend()
    axes[1,0].grid(True)
    
    # 4. Distribucin de probabilidades
    axes[1,1].hist(y_pred_proba[y_val == 0], bins=30, alpha=0.7, 
                   label='No Subscription', color='red')
    axes[1,1].hist(y_pred_proba[y_val == 1], bins=30, alpha=0.7, 
                   label='Subscription', color='blue')
    axes[1,1].set_xlabel('Probabilidad Predicha')
    axes[1,1].set_ylabel('Frecuencia')
    axes[1,1].set_title('Distribucin de Probabilidades')
    axes[1,1].legend()
    
    plt.tight_layout()
    plt.show()

# Llamar la funcin de visualizacin
visualizar_metricas_completas(y_valid, y_pred, y_pred_proba)
```
